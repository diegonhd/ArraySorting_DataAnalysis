# Gráfico de Linhas e Análise de Desempenho

## `Bubble Sort`

```python
inttamanho = [1000, 10000, 20000, 30000, 40000, 50000]

fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_bs["Média dos Aleatórios"],
        label = "Média dos Aleatórios",
        color = 'blue',
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_bs["Pior Caso"],
        label = "Pior Caso",
        color = 'red',
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_bs["Aleatório 1"],
        label = "Aleatório 1",
        color = 'yellow',
        linewidth = 1,)

ax.plot(inttamanho, df_bs["Aleatório 2"],
        label = "Aleatório 2",
        color = 'green',
        linewidth = 1,)

ax.plot(inttamanho, df_bs["Aleatório 3"],
        label = "Aleatório 3",
        color = 'orange',
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Bubble Sort)")
ax.legend()
ax.grid(True)
```

### Como se observa no gráfico acima, todas as curvas são exponenciais, complexidade temporal que já é conhecida observando o próprio algoritmo de ordenação.


### Se vê também que a curva é mais acentuada na função de pior caso, o que mostra uma má estabilidade no método.

### Então, a complexidade tanto de Caso Médio quanto de Pior Caso é de:
### $$C(t) = O(n^2)$$

# `Selection Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_ss["Média dos Aleatórios"],
        label = "Média dos Aleatórios",
        color = 'blue',
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_ss["Pior Caso"],
        label = "Pior Caso",
        color = 'red',
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_ss["Aleatório 1"],
        label = "Aleatório 1",
        color = 'yellow',
        linewidth = 1,)

ax.plot(inttamanho, df_ss["Aleatório 2"],
        label = "Aleatório 2",
        color = 'green',
        linewidth = 1,)

ax.plot(inttamanho, df_ss["Aleatório 3"],
        label = "Aleatório 3",
        color = 'orange',
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Selection Sort)")
ax.legend()
ax.grid(True)
```

## `Shell Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_sh["Média dos Aleatórios"],
        label = "Média dos Aleatórios",
        color = 'blue',
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_sh["Pior Caso"],
        label = "Pior Caso",
        color = 'red',
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_sh["Aleatório 1"],
        label = "Aleatório 1",
        color = 'yellow',
        linewidth = 1,)

ax.plot(inttamanho, df_sh["Aleatório 2"],
        label = "Aleatório 2",
        color = 'green',
        linewidth = 1,)

ax.plot(inttamanho, df_sh["Aleatório 3"],
        label = "Aleatório 3",
        color = 'orange',
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Shell Sort)")
ax.legend()
ax.grid(True)
```

### Algo interessante aconteceu nesse gráfico: o desempenho do 'pior caso' foi melhor que a do caso médio. Provavelmente pelo fato de que para a estratégia de gap original usada - gap = n//2, n//4, n//8... - o caso de pior desempenho não é o array inversamente ordenado. Além disso, o array decrescente pode ter 'facilitado' a ordenação.

### Analisando apenas o comportamento das curvas, observa-se uma maior linearidade das curvas. Mas, apesar de aparentemente não ser, é conhecido que a complexidade de ambos os casos para essa estratégia de gap também é de: $$ C(t) = O(n^2) $$ o que mostra que o desempenho prático é bem melhor que o teórico, ainda mais se vendo a escalabilidade do gráfico (0s - 0.5s).

## `Quick Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_qs["Média dos Aleatórios"],
        label = "Média dos Aleatórios",
        color = 'blue',
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_qs["Pior Caso"],
        label = "Pior Caso",
        color = 'red',
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_qs["Aleatório 1"],
        label = "Aleatório 1",
        color = 'yellow',
        linewidth = 1,)

ax.plot(inttamanho, df_qs["Aleatório 2"],
        label = "Aleatório 2",
        color = 'green',
        linewidth = 1,)

ax.plot(inttamanho, df_qs["Aleatório 3"],
        label = "Aleatório 3",
        color = 'orange',
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Quick Sort)")
ax.legend()
ax.grid(True)
```

### O Quick Sort teve o melhor desempenho até esse momento. O array inversamente ordenado teve um desempenho péssimo e a curva tem um comportamento parecido com uma curva quadrática:
### $$ C(t) = O(n ^ 2) $$

### Apesar disso, o tempo dos casos médios foi extremamente baixo, o que demonstra o quão eficiente é o Quick Sort no caso médio:
### $$ C(t) = O(n log (n)) $$

## `Merge Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_ms["Média dos Aleatórios"],
        label = "Média dos Aleatórios",
        color = 'blue',
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_ms["Pior Caso"],
        label = "Pior Caso",
        color = 'red',
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_ms["Aleatório 1"],
        label = "Aleatório 1",
        color = 'yellow',
        linewidth = 1,)

ax.plot(inttamanho, df_ms["Aleatório 2"],
        label = "Aleatório 2",
        color = 'green',
        linewidth = 1,)

ax.plot(inttamanho, df_ms["Aleatório 3"],
        label = "Aleatório 3",
        color = 'orange',
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Merge Sort)")
ax.legend()
ax.grid(True)
```

### Parecido com o caso do Shell Sort, o 'Pior Caso' demonstrou na verdade ser melhor que o 'Caso Médio', mas com uma diferença: os casos estão muito equiparados, o que faz-se concluir que esse método trabalha em uma escala semelhante ao do Quick Sort, porém, apresenta uma grande consistencia em tempo.

### Os dois casos então apresentam complexidade:
### $$ C(t) = O(n log (n)) $$

### O Merge Sort alcança essa estabilidade de tempo $O(n \log n)$ ao custo de espaço. As linhas abaixo (que estão em `arrays.ipynb`) criam cópias dos subarrays, o que significa que o seu algoritmo usa $O(n)$ de espaço extra na memória.

### `esquerda = lista[inicio:meio].copy()`


### `direita = lista[meio:fim].copy()`

### O que dá pra tirar de insight dessa comparação é que existe claramente uma subdivisão de desempenho entre os métodos.

### Os 3 métodos mais ao topo (Bubble, Selection e Insertion) estão em uma escala 10² segundos enquanto que o Shell, Quick e Merge estão em uma escala menor ou igaul a 10^-1 segundos, com destaque ao Quick Sort, que trabalha em  10^-2  e, logo, se mostrou ser aproximadamente 10x mais rápido que os outros dois quando se fala de caso médio. Isso demonstra claramente a diferença de complexidade discutida anteriormente entre $$C(t) = O(n^2)$$ $$e$$ $$C(t) = O(nlog(n)).$$

### Nessa comparação ocorreu uma mudança brusca em relação ao Quick Sort. Na comparação anterior (entre casos médios), o Quick Sort se isolou como o melhor desempenho, mas, agora, quando se compara ele com o Merge Sort e o Shell Sort, que possuíam a mesma ordem, observa se que em vez de ele ser 10x mais rápido, foi 10x mais lento.

### Isso evidencia que o Quick Sort, neste caso, voltou a ter complexidade de escala quadrática $C(t) = O(n^2)$.

### Quando vemos o comportamento do Merge Sort e do Shell Sort mantiveram a mesma ordem, o que demonstra que ambos realmente garantem um desempenho $C(t) = O(nlog(n))$.

### O que dá pra tirar de insight dessa comparação é que existe claramente uma subdivisão de desempenho entre os métodos.

### Os 3 métodos mais ao topo (Bubble, Selection e Insertion) estão em uma escala 10² segundos, o que é esperado considerando que são algoritmos de complexidade quadrática

### No meio termo, temos o Shell Sort, que tem um desempenho intermediário, mas melhor que os algoritmos de complexidade $O$ = $n$²

### Na parte de baixo do gráfico, temos os algoritmos de melhores performances em caso médio, com complexidade  $O = nlog(n)$ ou $O = n + k$, em que 'k' é o range dos números (caso do bucket, radix e counting).

```python
inttamanho = [1000, 10000, 20000, 30000, 40000, 50000]

# --- Estilos de Linha e Cores por Complexidade ---
estilos = {
    'O(n^2)': {'linestyle': ':', 'linewidth': 2},
    'O(n log n)': {'linestyle': '--', 'linewidth': 3},
    'O(n+k)': {'linestyle': '-', 'linewidth': 4},
}

# Mapeamento de Algoritmos para Grupos de Complexidade
mapeamento = {
    'Bubble Sort': 'O(n^2)',
    'Selection Sort': 'O(n^2)',
    'Insertion Sort': 'O(n^2)',
    'Shell Sort': 'O(n log n)', # Mantido em O(n log n) por ser o melhor grupo subsequente
    'Quick Sort': 'O(n log n)',
    'Merge Sort': 'O(n log n)',
    'Heap Sort': 'O(n log n)',
    'Counting Sort': 'O(n+k)',
    'Radix Sort': 'O(n+k)',
    'Bucket Sort': 'O(n+k)',
}

fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')

for sort_method, complexidade_grupo in mapeamento.items():
    style = estilos[complexidade_grupo]
    
    # Define a cor com base no algoritmo, pode personalizar mais se quiser
    if sort_method == 'Quick Sort':
        color = 'red'
    elif sort_method == 'Counting Sort':
        color = 'green'
    elif sort_method == 'Bubble Sort':
        color = 'blue'
    else:
        # Pega a cor base do grupo
        if complexidade_grupo == 'O(n^2)':
            color = 'lightcoral'
        elif complexidade_grupo == 'O(n log n)':
            color = 'darkorange'
        elif complexidade_grupo == 'O(n+k)':
            color = 'darkgreen'
    
    ax.plot(inttamanho, df_medias[sort_method], 
            label=f'{sort_method} ({complexidade_grupo})',
            color=color,
            linestyle=style['linestyle'],
            linewidth=style['linewidth'])

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do Array (n)', fontsize=12)
ax.set_ylabel('Média de Tempo (s) - Escala Logarítmica', fontsize=12)
ax.set_title('Comparação de Algoritmos de Ordenação (Caso Médio)', fontsize=14)

# Define a escala do eixo Y como logarítmica
ax.set_yscale('log')

# Adiciona linhas de grade em ambas as escalas
ax.grid(True, which="both", ls="--", alpha=0.7)

# Adiciona legenda com bom posicionamento
ax.legend(loc='lower right', bbox_to_anchor=(1.0, 0.0), fontsize=9)

plt.show()
```
