## Estrutura de Índices para `list_x` e `list_arr_(n)k`:


### `list_x`, onde x é a abreviação do método de ordenamento:
* **Índices `[0]` a `[6]`**: Representam as listas geradas pelos tempos relacionados a cada tamanho de array em progressão do método x
(1k, 10k, 20k, 30k, 40k, 50k).
---------------------------------------------------------------------------------------------------------------------------------------
### `list_arr_(n)k`, onde `(n)k` representa o tamanho do array (ex: 10k, 50k, 100k):

* **Índices `[0]` a `[2]`: Tempos de Execução (Arrays Aleatórios)**
    * **`[0]`:** Tempo da 1ª execução aleatória.
    * **`[1]`:** Tempo da 2ª execução aleatória.
    * **`[2]`:** Tempo da 3ª execução aleatória.

* **Índice `[3]`: Tempo de Execução (Pior Caso)**
    * Representa o tempo de execução para o array ordenado no pior caso (geralmente decrescente para algoritmos de ordenação).

* **Índice `[4]`: Média dos Tempos Aleatórios**
    * Valor médio calculado a partir dos tempos nos índices `[0]`, `[1]`, e `[2]`.
---------------------------------------------------------------------------------------------------------------------------------------

# Gráfico de Linhas e Análise de Desempenho

## `Bubble Sort`

```python
inttamanho = [1000, 10000, 20000, 30000, 40000, 50000]

fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_bs["Média dos Aleatórios"],
        label = "Média dos Aleatórios", 
        color = 'blue', 
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_bs["Pior Caso"],
        label = "Pior Caso", 
        color = 'red', 
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_bs["Aleatório 1"],
        label = "Aleatório 1", 
        color = 'yellow', 
        linewidth = 1,)

ax.plot(inttamanho, df_bs["Aleatório 2"],
        label = "Aleatório 2", 
        color = 'green', 
        linewidth = 1,)

ax.plot(inttamanho, df_bs["Aleatório 3"],
        label = "Aleatório 3", 
        color = 'orange', 
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Bubble Sort)")
ax.legend()
ax.grid(True)
```

### Como se observa no gráfico acima, todas as curvas são exponenciais, complexidade temporal que já é conhecida observando o próprio algoritmo
### de ordenação.


### Se vê também que a curva é mais acentuada na função de pior caso, o que mostra uma má estabilidade no método.

### Então, a complexidade tanto de Caso Médio quanto de Pior Caso é de:
### $$C(t) = O(n^2)$$

# `Selection Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_ss["Média dos Aleatórios"],
        label = "Média dos Aleatórios", 
        color = 'blue', 
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_ss["Pior Caso"],
        label = "Pior Caso", 
        color = 'red', 
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_ss["Aleatório 1"],
        label = "Aleatório 1", 
        color = 'yellow', 
        linewidth = 1,)

ax.plot(inttamanho, df_ss["Aleatório 2"],
        label = "Aleatório 2", 
        color = 'green', 
        linewidth = 1,)

ax.plot(inttamanho, df_ss["Aleatório 3"],
        label = "Aleatório 3", 
        color = 'orange', 
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Selection Sort)")
ax.legend()
ax.grid(True)
```

### Ao contrário do Bubble Sort, o comportamento de todas as funções foram semelhantes, além de suportar uma escala bem menor de tempo (0s, 200s) x (0s, 700s).

### Mas, existe uma grande semelhança com o método anterior: os dois tem curvas que se comportam da mesma maneira (exponencial).

### Esse fato implica que a complexidade desse método para o Caso Médio e Pior Caso também é de:
### $$C(t) = O(n^2)$$

## `Insertion Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_ins["Média dos Aleatórios"],
        label = "Média dos Aleatórios", 
        color = 'blue', 
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_ins["Pior Caso"],
        label = "Pior Caso", 
        color = 'red', 
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_ins["Aleatório 1"],
        label = "Aleatório 1", 
        color = 'yellow', 
        linewidth = 1,)

ax.plot(inttamanho, df_ins["Aleatório 2"],
        label = "Aleatório 2", 
        color = 'green', 
        linewidth = 1,)

ax.plot(inttamanho, df_ins["Aleatório 3"],
        label = "Aleatório 3", 
        color = 'orange', 
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Insertion Sort)")
ax.legend()
ax.grid(True)
```

### Para o Insertion Sort, vê-se que os casos se comportam que nem o do Bubble Sort, mas em uma escala menor que a dele (0s-500s) x (0s-700s), porém em uma escala maior do que a do Selection Sort.

### O que se destaca nesse método é que sua complexidade para o melhor caso (array já ordenado) é a melhor possível:
### $$ C(t) = O(n)

## `Shell Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_sh["Média dos Aleatórios"],
        label = "Média dos Aleatórios", 
        color = 'blue', 
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_sh["Pior Caso"],
        label = "Pior Caso", 
        color = 'red', 
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_sh["Aleatório 1"],
        label = "Aleatório 1", 
        color = 'yellow', 
        linewidth = 1,)

ax.plot(inttamanho, df_sh["Aleatório 2"],
        label = "Aleatório 2", 
        color = 'green', 
        linewidth = 1,)

ax.plot(inttamanho, df_sh["Aleatório 3"],
        label = "Aleatório 3", 
        color = 'orange', 
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Shell Sort)")
ax.legend()
ax.grid(True)
```

### Algo interessante aconteceu nesse gráfico: o desempenho do 'pior caso' foi melhor que a do caso médio. Provavelmente pelo fato de que para a estratégia de gap original usada - gap = n//2, n//4, n//8... - o caso de pior desempenho não é o array inversamente ordenado. Além disso, o array decrescente pode ter 'facilitado' a ordenação. 

### Analisando apenas o comportamento das curvas, observa-se uma maior linearidade das curvas. Mas, apesar de aparentemente não ser, é conhecido que a complexidade de ambos os casos para essa estratégia de gap também é de: $$ C(t) = O(n^2) $$ o que mostra que o desempenho prático é bem melhor que o teórico, ainda mais se vendo a escalabilidade do gráfico (0s - 0.5s).

## `Quick Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_qs["Média dos Aleatórios"],
        label = "Média dos Aleatórios", 
        color = 'blue', 
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_qs["Pior Caso"],
        label = "Pior Caso", 
        color = 'red', 
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_qs["Aleatório 1"],
        label = "Aleatório 1", 
        color = 'yellow', 
        linewidth = 1,)

ax.plot(inttamanho, df_qs["Aleatório 2"],
        label = "Aleatório 2", 
        color = 'green', 
        linewidth = 1,)

ax.plot(inttamanho, df_qs["Aleatório 3"],
        label = "Aleatório 3", 
        color = 'orange', 
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Quick Sort)")
ax.legend()
ax.grid(True)
```

### O Quick Sort teve o melhor desempenho até esse momento. O array inversamente ordenado teve um desempenho péssimo e a curva tem um comportamento parecido com uma curva quadrática:
### $$ C(t) = O(n ^ 2) $$

### Apesar disso, o tempo dos casos médios foi extremamente baixo, o que demonstra o quão eficiente é o Quick Sort no caso médio:
### $$ C(t) = O(n log (n)) $$

## `Merge Sort`

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_ms["Média dos Aleatórios"],
        label = "Média dos Aleatórios", 
        color = 'blue', 
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_ms["Pior Caso"],
        label = "Pior Caso", 
        color = 'red', 
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_ms["Aleatório 1"],
        label = "Aleatório 1", 
        color = 'yellow', 
        linewidth = 1,)

ax.plot(inttamanho, df_ms["Aleatório 2"],
        label = "Aleatório 2", 
        color = 'green', 
        linewidth = 1,)

ax.plot(inttamanho, df_ms["Aleatório 3"],
        label = "Aleatório 3", 
        color = 'orange', 
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Merge Sort)")
ax.legend()
ax.grid(True)
```

### Parecido com o caso do Shell Sort, o 'Pior Caso' demonstrou na verdade ser melhor que o 'Caso Médio', mas com uma diferença: os casos estão muito equiparados, o que faz-se concluir que esse método trabalha em uma escala semelhante ao do Quick Sort, porém, apresenta uma grande consistencia em tempo.

### Os dois casos então apresentam complexidade:
### $$ C(t) = O(n log (n)) $$

### O Merge Sort alcança essa estabilidade de tempo $O(n \log n)$ ao custo de espaço. As linhas abaixo (que estão em arrays.ipynb) criam cópias dos subarrays, o que significa que o seu algoritmo usa $O(n)$ de espaço extra na memória.

### `esquerda = lista[inicio:meio].copy()`


### `direita = lista[meio:fim].copy()`
