```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Carregando os arquivos

```python
# Carregando as listas dos arquivos gerados por cada método
list_bs = np.load('array archive/list_bs.npy')
list_ss = np.load('array archive/list_ss.npy')
list_ins = np.load('array archive/list_ins.npy')
list_sh = np.load('array archive/list_sh.npy')
list_qs = np.load('array archive/list_qs.npy')
list_ms = np.load('array archive/list_ms.npy')
```

## Estrutura de Índices para `list_x` e `list_arr_(n)k`:


### `list_x`, onde x é a abreviação do método de ordenamento:
* **Índices `[0]` a `[6]`**: Representam as listas geradas pelos tempos relacionados a cada tamanho de array em progressão do método x
(1k, 10k, 20k, 30k, 40k, 50k).
---------------------------------------------------------------------------------------------------------------------------------------
### `list_arr_(n)k`, onde `(n)k` representa o tamanho do array (ex: 10k, 50k, 100k):

* **Índices `[0]` a `[2]`: Tempos de Execução (Arrays Aleatórios)**
    * **`[0]`:** Tempo da 1ª execução aleatória.
    * **`[1]`:** Tempo da 2ª execução aleatória.
    * **`[2]`:** Tempo da 3ª execução aleatória.

* **Índice `[3]`: Tempo de Execução (Pior Caso)**
    * Representa o tempo de execução para o array ordenado no pior caso (geralmente decrescente para algoritmos de ordenação).

* **Índice `[4]`: Média dos Tempos Aleatórios**
    * Valor médio calculado a partir dos tempos nos índices `[0]`, `[1]`, e `[2]`.
---------------------------------------------------------------------------------------------------------------------------------------

# Gráfico de Linhas e Análise de Desempenho

## `Bubble Sort`

```python
inttamanho = [1000, 10000, 20000, 30000, 40000, 50000]

fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(inttamanho, df_bs["Média dos Aleatórios"],
        label = "Média dos Aleatórios", 
        color = 'blue', 
        linewidth = 5,
        linestyle = ':')

ax.plot(inttamanho, df_bs["Pior Caso"],
        label = "Pior Caso", 
        color = 'red', 
        linewidth = 2,
        linestyle = '--')

ax.plot(inttamanho, df_bs["Aleatório 1"],
        label = "Aleatório 1", 
        color = 'yellow', 
        linewidth = 1,)

ax.plot(inttamanho, df_bs["Aleatório 2"],
        label = "Aleatório 2", 
        color = 'green', 
        linewidth = 1,)

ax.plot(inttamanho, df_bs["Aleatório 3"],
        label = "Aleatório 3", 
        color = 'orange', 
        linewidth = 1,)

ax.set_xlim(1000, 50000)
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempo (Bubble Sort)")
ax.legend()
ax.grid(True)
```
