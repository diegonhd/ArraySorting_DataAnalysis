```python
# Carregando as médias dos tempos
mean_time_1k = np.load('archive/mean_arr_1k.npy')
mean_time_5k = np.load('archive/mean_arr_5k.npy')
mean_time_10k = np.load('archive/mean_arr_10k.npy')
mean_time_20k = np.load('archive/mean_arr_20k.npy')
mean_time_30k = np.load('archive/mean_arr_30k.npy')
mean_time_40k = np.load('archive/mean_arr_40k.npy')
mean_time_50K = np.load('archive/mean_arr_50k.npy')
```

## Média x Pior Caso Em Cada 'N' (Tamanho Da Lista)

```python
y = np.hstack(([mean_time_1k, time_wc_1k], list_arr_1k))
cores = ['blue', 'red', 'green', 'orange', 'purple']
categorias = np.array(['Média', 'Pior Caso', 'Aleatório 1', 'Aleatório 2', 'Aleatório 3'])
fig, ax = plt.subplots(figsize = (5, 2.7), layout = 'constrained')
ax.bar(categorias, y, color = cores)
ax.set_ylabel('Tempo(s)')
ax.set_title("Comparação interna entre os arrays de 1k")
```

### Gráfico de Linhas 

```python
list_len = np.array([1000, 5000, 10000,20000, 30000, 40000, 50000])
mean_times = np.array([mean_time_1k, mean_time_5k, mean_time_10k, mean_time_20k, mean_time_30k, mean_time_40k, mean_time_50K])
worst_case_times = np.array([time_wc_1k, time_wc_5k, time_wc_10k, time_wc_20k, time_wc_30k, time_wc_40k, time_wc_50k])
rand_arr1 = np.array([list_arr_1k[0], list_arr_5k[0], list_arr_10k[0], list_arr_20k[0], list_arr_30k[0], list_arr_40k[0], list_arr_50k[0]])
rand_arr2 = np.array([list_arr_1k[1], list_arr_5k[1], list_arr_10k[1], list_arr_20k[1], list_arr_30k[1], list_arr_40k[1], list_arr_50k[1]])
rand_arr3 = np.array([list_arr_1k[2], list_arr_5k[2], list_arr_10k[2], list_arr_20k[2], list_arr_30k[2], list_arr_40k[2], list_arr_50k[2]])
```

```python
fig, ax = plt.subplots(figsize = (7,4), layout = 'constrained')

ax.plot(list_len, mean_times, 
        label= 'Médias dos tempos',
        color='blue',
        linewidth = 5,
        linestyle = ':')

ax.plot(list_len, worst_case_times,
        label= 'Piores tempos possíveis',
        color='red',
        linestyle = '--',
        linewidth = 2)

ax.plot(list_len, rand_arr1,
        label = 'Linha Aleatória 1',
        color = 'yellow',
        linewidth = 1)

ax.plot(list_len, rand_arr2,
        label = 'Linha Aleatória 2',
        color = 'black',
        linewidth = 1)

ax.plot(list_len, rand_arr3,
        label = 'Linha Aleatória 3',
        color = 'orange',
        linewidth = 1)

ax.set_xlim(1000, 50000) 
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempo (s)')
ax.set_title("Tamanho x Tempos")
ax.legend()
ax.grid(True)
'''is not a valid value for ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'''
```
