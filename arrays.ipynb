```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# Definindo a função bubble sort sem lib (obrigatório)
def bubble_sort(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista
```

```python
# Definindo a função de retorno do tempo que o array levou a ser ordenado
def tempo_ordenaçao(lista):
    start_time = time.time()
    bubble_sort(lista)
    end_time = time.time()
    total_time = end_time - start_time
    return total_time
```

## Para o caso dos arrays de size = 1k

```python
# Gerando os 3 arrays aleatórios de cada caso (1k)
arr1_1k = np.random.randint(1,1001,1000)
arr2_1k = np.random.randint(1,1001,1000)
arr3_1k = np.random.randint(1,1001,1000)
arr_1k_sorted = np.random.randint(1,1001,1000)

# Fazendo um array do pior caso 
arr_wc_1k = arr_1k_sorted.sort()
arr_wc_1k = arr1_1k[::-1]
```

```python
# Pegando o tempo de ordenação dos 4 arrays
time_arr1_1k =  tempo_ordenaçao(arr1_1k)
print(f"Tempo do primeiro array de 1k: {time_arr1_1k} segundos")
time_arr2_1k = tempo_ordenaçao(arr2_1k)
print(f"Tempo do primeiro array de 1k: {time_arr2_1k} segundos")
time_arr3_1k = tempo_ordenaçao(arr3_1k)
print(f"Tempo do primeiro array de 1k: {time_arr3_1k} segundos")
time_wc_1k = tempo_ordenaçao(arr_wc_1k)
print(f"Tempo do primeiro array de 1k: {time_wc_1k} segundos")
```

```python
# Fazendo uma média dos 3 casos dos arrays aleatórios
list_arr_1k = np.array([time_arr1_1k, time_arr2_1k, time_arr3_1k])
mean_arr_1k = list_arr_1k.mean()
print(f"Tempo médio de ordenação dos arrays de 1k {mean_arr_1k} segundos.")
```
